# Classe mère

```python
class Personnage():

    """ Classe mère qui définit un personnage """

    def __init__(self, nom, vie, vitesse, esquive):
        self.nom = nom
        self.vie = vie
        self.vitesse = vitesse
        self.esquive = esquive
        self.is_pepsie = False

    @property
    def nom(self):
        return self._nom
    
    @nom.setter
    def nom(self, nom):
        self._nom = nom

    @property
    def vie(self):
        return self._vie
    
    @vie.setter
    def vie(self, vie):
        self._vie = vie

    @property
    def vitesse(self):
        return self._vitesse

    @vitesse.setter
    def vitesse(self, vitesse):
        self._vitesse = vitesse

    @property
    def esquive(self):
        return self._esquive

    @esquive.setter
    def esquive(self, esquive):
        self._esquive = esquive

    def is_pepsie(self):
        """ Renvoie True si le personnage est pepsie """
        return self.vie <= 0.

    def __str__(self):
        return f"{self.nom} a {self.vie} points de vie, une vitesse de {self.vitesse} et une esquive de {self.esquive}."

perso_test = Personnage("Benoit", 110, 0, 10)
print(perso_test)
```

```python
class Guerrier(Personnage):

    """ Classe fille qui définit un guerrier """

    def __init__(self, nom, vie, vitesse, esquive, force):
        super().__init__(nom, vie, vitesse, esquive)
        self.force = force

    @property
    def force(self):
        return self._force
    
    @force.setter
    def force(self, force):
        self._force = force

    def attaque(self, cible):
        """ Méthode qui permet d'attaquer une cible """
        print("{} attaque {}".format(self.nom, cible.nom))
        cible.vie -= self.force
        print("{} a perdu {} points de vie il lui reste {}".format(cible.nom, self.force, cible.vie))
        if cible.vie <= 0:
            print("{} est mort".format(cible.nom))
            cible.is_pepsie = True

    def __str__(self):
        return "Guerrier \n "+super().__str__() +"\n-- Force: {}".format(self.force)

perso_test = Guerrier("Test", 100, 10, 10, 10)
print(perso_test)
```

```python
class Magicien(Personnage):

    """ Classe fille qui définit un magicien """

    def __init__(self, nom, vie, vitesse, esquive, mana):
        super().__init__(nom, vie, vitesse, esquive)
        self.mana = mana

    @property
    def mana(self):
        return self._mana
    
    @mana.setter
    def mana(self, mana):
        self._mana = mana

    def lancer_sort(self, cible):
        """ Méthode qui permet de lancer un sort sur une cible """
        print("{} lance un sort sur {}".format(self.nom, cible.nom))
        cible.vie -= self.mana
        print("{} a perdu {} points de vie il lui reste {}".format(cible.nom, self.mana, cible.vie))
        if cible.vie <= 0:
            print("{} est mort".format(cible.nom))
            cible.is_pepsie = True

    def __str__(self):
        return "Magicien \n"+ super().__str__() + " \n-- Mana: {}".format(self.mana)

perso_test = Magicien("Test", 100, 10, 10, 10)
print(perso_test)
```

```python
class Archer(Personnage):

    """ Classe fille qui définit un archer """

    def __init__(self, nom, vie, vitesse, esquive, precision):
        super().__init__(nom, vie, vitesse, esquive)
        self.precision = precision

    @property
    def precision(self):
        return self._precision
    
    @precision.setter
    def precision(self, precision):
        self._precision = precision

    def tirer(self, cible):
        """ Méthode qui permet de tirer sur une cible """
        print("{} tire sur {}".format(self.nom, cible.nom))
        cible.vie -= self.precision
        print("{} a perdu {} points de vie il lui reste {}".format(cible.nom, self.precision, cible.vie))
        if cible.vie <= 0:
            print("{} est mort".format(cible.nom))
            cible.is_pepsie = True

    def __str__(self):
        return "Archer \n"+ super().__str__()+" \n-- Precision: {}".format(self.precision)

perso_test = Archer("Test", 100, 10, 10, 10)
print(perso_test)
```

# Le bon, la brute et le truand

##### Création des personnages

```python
Le_Guerrier = Guerrier("Le Guerrier", 100, 10, 10, 10)
Le_Magicien = Magicien("Le Magicien", 100, 10, 10, 10)
L_Archer = Archer("L'Archer", 100, 10, 10, 10)
```

# Le Jeu

```python
Le_Guerrier = Guerrier("Le Guerrier", 100, 10, 10, 10)
Le_Magicien = Magicien("Le Magicien", 100, 10, 10, 10)
L_Archer = Archer("L'Archer", 100, 10, 10, 10)

from numpy import random
while True:
    # Le jeu s'arrete si deux des personnages sont morts
    le_test_is_pepsie = Le_Guerrier.is_pepsie + Le_Magicien.is_pepsie + L_Archer.is_pepsie
    if le_test_is_pepsie == 2:
        break
    # on test la rapidité des joueurs 
    if Le_Guerrier.vitesse > Le_Magicien.vitesse and Le_Guerrier.vitesse > L_Archer.vitesse:
        # on choisi une cible aléatoire
        cible = random.choice([Le_Magicien, L_Archer])
        # il attaque la cible
        Le_Guerrier.attaque(cible)
    elif Le_Magicien.vitesse > Le_Guerrier.vitesse and Le_Magicien.vitesse > L_Archer.vitesse:
        # on choisi une cible aléatoire
        cible = random.choice([Le_Guerrier, L_Archer])
        # il attaque la cible
        Le_Magicien.lancer_sort(cible)
    elif L_Archer.vitesse > Le_Guerrier.vitesse and L_Archer.vitesse > Le_Magicien.vitesse:
        # on choisi une cible aléatoire
        cible = random.choice([Le_Guerrier, Le_Magicien])
        # il attaque la cible
        L_Archer.tirer(cible)
    else:
        print("Egalité")

    print("\n")
    # on change la vitesse des joueurs aléatoirement
    Le_Guerrier.vitesse = Le_Guerrier.vitesse + random.randint(1, 10) - 5
    Le_Magicien.vitesse = Le_Magicien.vitesse + random.randint(1, 10) - 5
    L_Archer.vitesse = L_Archer.vitesse + random.randint(1, 10) - 5
    
if(Le_Guerrier.is_pepsie):
    print("Le Guerrier est mort")
else:
    print("Le Guerrier a gagné")

if(Le_Magicien.is_pepsie):
    print("Le Magicien est mort")
else:
    print("Le Magicien a gagné")    

if(L_Archer.is_pepsie):
    print("L'Archer est mort")
else:
    print("L'Archer a gagné")
```
